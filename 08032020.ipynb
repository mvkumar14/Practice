```python
def lat(s):
    a = s.split(",")
    return float(a[1][:-1])
def long(s):
    a = s.split(",")
    return float(a[0][1:])
```

```python
df['lattitude'] = df.location.apply(lat)
```

```python
df['longitude'] = df.location.apply(long)
```

```python
def square_selection(df,center_long,center_lat,square_size=4):
    diff = square_size/2
    condition = df['longitude'] > center_long - diff
    condition2 = df['longitude'] < center_long + diff
    condition3 = df['lattitude'] > center_lat - diff
    condition4 =  df['lattitude'] < center_lat + diff
    return df[condition]
```
